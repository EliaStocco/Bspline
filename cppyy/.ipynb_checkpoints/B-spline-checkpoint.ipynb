# Funzioni di base 1D

Imposto la cardinalit√† della base, il grado polinomiale e gli "knots"

```python
N=3 #numero di funzioni di base
P=2 #grado polinomiale
v=vector['double']([0,1,2,3,4,5])
list(v)
```

Definisco il knot vector

```python
kv=Bspline.knot_vector['double'](P,N,v)
```

```python
kv_v=Bspline.Bspline['double'].knots_t()#array di knot vectors
kv_v[0]=kv
```

Definisco una classe contenente le funzioni di base per le B-spline

```python
Bs=Bspline.Bspline['double'](kv_v)
```

Alloco una B-spline, i cui coefficienti saranno riferiti rispetto alla base appena definita. 

```python
N_x=1000
x  =np.linspace(min(v),max(v),N_x,endpoint=False)
x_std=vector[Bspline.Bspline['double'].Type_in](N_x)

for i in range(0,N_x):
    x_std[i][0]=x[i]
```

```python
index=array['int',1]()
coeff=array['double',1]()

plt.figure ( 0 , figsize = ( 15 , 5 ) )
colors = iter(cm.rainbow(np.linspace(0, 1,N)))

for i in range(0,N):
    #print(i)
    Bs.clear()
    index[0]=i
    coeff[0]=1.0
    Bs.set(index,coeff)
    y_std=Bs(x_std)
    y=np.zeros(N_x,dtype=float)
    for j in range(0,N_x):
        y[j]=y_std[j][0]
    
    plt.plot(x, y, color= next(colors),label=str(i))
    
plt.legend()
plt.grid(True)
plt.show()
```

## Gradiente

```python
N=3 #numero di funzioni di base
P=2 #grado polinomiale
v=vector['double']([0,1,2,3,4,5])
list(v)

#
kv=Bspline.knot_vector['double'](P,N,v)

#
kv_v=Bspline.Bspline['double'].knots_t()#array di knot vectors
kv_v[0]=kv

#
Bs=Bspline.Bspline['double'](kv_v)
```

```python
N_x=1000
x  =np.linspace(min(v),max(v),N_x,endpoint=False)
x_std=vector[Bspline.Bspline['double'].Type_in](N_x)

for i in range(0,N_x):
    x_std[i][0]=x[i]
```

```python
index=array['int',1]()
coeff=array['double',1]()

plt.figure ( 0 , figsize = ( 15 , 5 ) )
colors = iter(cm.rainbow(np.linspace(0, 1,N)))

for i in range(0,N):
    #print(i)
    Bs.clear()
    index[0]=i
    coeff[0]=1.0
    Bs.set(index,coeff)
    y_std=Bs(x_std)
    grad=Bs.jacobian(x_std)
    y=np.zeros(N_x,dtype=float)
    for j in range(0,N_x):
        y[j]=grad[0][j][0]
    
    plt.plot(x, y, color= next(colors),label=str(i))
    
plt.legend()
plt.grid(True)
plt.show()
```

```python
Bs=Bspline.Bspline['double'](kv_v)
```

```python
N_x=1000
x  =np.linspace(min(v),max(v),N_x,endpoint=False)
x_std=vector[Bspline.Bspline['double'].Type_in](N_x)

for i in range(0,N_x):
    x_std[i][0]=x[i]
```

# Dominio in $\mathbb{R}^2$

```python
P=1 #grado polinomiale
v1=np.asarray(np.arange(0,4))
```

```python
begin=np.zeros(P)
end=np.zeros(P)
begin.fill(min(v1))
end.fill(max(v1))
v_py=np.concatenate((begin, v1, end), axis=0)
v=vector['double'](v_py)
print(v_py)
```

```python
N=len(v)-P-1
#
kv=Bspline.knot_vector['double'](P,N,v)
kv_v=Bspline.Bspline['double',2,2].knots_t()#array di knot vectors
kv_v[0]=kv
kv_v[1]=kv
```

```python
Bs=Bspline.Bspline['double',2,2](kv_v)
```

```python
N_x=50
N_y=50
x  =np.linspace(min(v),max(v),N_x,endpoint=False)
y  =np.linspace(min(v),max(v),N_y,endpoint=False)
X, Y = np.meshgrid(x, y)

Y_vect=Y.ravel()
X_vect=X.ravel()

XY = vector[array['double',2]](N_x*N_y)

for i in range(0,len(XY)):
    XY[i][0]=X_vect[i]
    XY[i][1]=Y_vect[i]  
```

```python
affine=[[1,2],\
       [3,4]]
```

```python
np.dot(affine,[1,1])
```

```python
a=np.dot(affine,[i,j])
```

```python
cp_x=np.zeros([N,N])
cp_y=np.zeros([N,N])
#matrice per un cambio di coordinate affine
affine=[[1,0],\
       [0,1]]
A=2
for i in range(0,N):
    for j in range(0,N):
        I=(i-1.5)/1.5
        J=(j-1.5)/1.5
        #a=np.dot(affine,[i,j])
        cp_x[i,j]=I
        cp_y[i,j]=(-A*pow(I,2)+2)*J
        #cp_x[i,j]=i
        #cp_y[i,j]=j
```

```python
index=array['int',2]()
coeff=array['double',2]()

k=1
for i in range(0,N):
    for j in range(0,N):
        #print(k)
        #ax = fig.add_subplot(N, N, k, projection='3d')
        k=k+1
        
        index[0]=i
        index[1]=j
        
        #setto le coordinate dei vari control points
        coeff[0]=cp_x[i,j]
        coeff[1]=cp_y[i,j]
        #coeff[2]=cp[i,j]
        
        #imposto il control point
        Bs.set(index,coeff)
        
```

```python
Z_stl=Bs(XY)
G=Bs.jacobian(XY)
J_stl=Bs.jacobian_determinant(XY)

Z_array=np.array(Z_stl) 
J_array=np.array(J_stl) 

X=Z_array[:,0].reshape(N_x,N_y)
Y=Z_array[:,1].reshape(N_x,N_y)
Z=J_array.reshape(N_x,N_y)
my_col=cm.jet(Z/np.amax(Z))

ZZ= np.array(X)
ZZ.fill(0)
```

```python
fig=plt.figure ( 0 , figsize = ( 10 , 8 ) )
fig.suptitle("2D domain P=" + str(P), fontsize=12 )
colors = iter(cm.rainbow(np.linspace(0, 1,N)))

ax = fig.add_subplot(1, 1, 1 , projection='3d')
#ax.scatter(cp_x,cp_y,color="red")

ax.plot_surface(X, Y, ZZ , facecolors = my_col , cmap='viridis',rstride=1, cstride=1)
ax.view_init(azim=0, elev=90)
plt.show()
```

```python
P=1#grado polinomiale
v1=np.asarray(np.arange(0,3))
```

```python
begin=np.zeros(P)
end=np.zeros(P)
begin.fill(min(v1))
end.fill(max(v1))
v_py=np.concatenate((begin, v1, end), axis=0)
v=vector['double'](v_py)
print(v_py)
```

```python
N=len(v)-P-1
#
kv=Bspline.knot_vector['double'](P,N,v)
kv_v=Bspline.Bspline['double',2,1].knots_t()#array di knot vectors
kv_v[0]=kv
kv_v[1]=kv
```

```python
Bs=Bspline.Bspline['double',2,1](kv_v)
```

```python
N_x=50
N_y=50
x  =np.linspace(min(v),max(v),N_x,endpoint=False)
y  =np.linspace(min(v),max(v),N_y,endpoint=False)
X, Y = np.meshgrid(x, y)

Y_vect=Y.ravel()
X_vect=X.ravel()

XY = vector[array['double',2]](N_x*N_y)

for i in range(0,len(XY)):
    XY[i][0]=X_vect[i]
    XY[i][1]=Y_vect[i]  
```

```python
index=array['int',2]()
coeff=array['double',1]()

fig=plt.figure ( 0 , figsize = ( 15 , 10 ) )
fig.suptitle("Bspline: 2D basis functions with polynomial degree P=" + str(P), fontsize=12 )
colors = iter(cm.rainbow(np.linspace(0, 1,N)))

Grad_x=np.zeros(len(XY))
Grad_y=np.zeros(len(XY))

k=1
for i in range(0,N):
    for j in range(0,N):
        #print(k)
        ax = fig.add_subplot(N, N, k, projection='3d')
        k=k+1
        
        Bs.clear()
        index[0]=i
        index[1]=j
        coeff[0]=1.0
        Bs.set(index,coeff)
        
        Z_stl=Bs(XY)
        grad =Bs.jacobian(XY)
        
        for ii in range(0,len(grad)):
            Grad_x[ii]=grad[ii][0][0]
            Grad_y[ii]=grad[ii][0][1]
            
        Z_array=np.array(Z_stl)        
        Z=Z_array.reshape(N_x,N_y)
        
        G=Grad_x.reshape(N_x,N_y)
        
        #plt.quiver(X_vect,Y_vect,Grad_x,Grad_y,angles='xy', scale_units='xy', scale=1)  
        
        ax.plot_surface(X, Y, G, cmap='viridis',rstride=1, cstride=1)
        #ax.set_title("("+str(i)+","+str(j)+")\n")


plt.show()
fig.savefig("images/Bspline-x-derivative-P="+str(P)+".pdf")  
fig.savefig("images/Bspline-x-derivative-P="+str(P)+".png")  
```
