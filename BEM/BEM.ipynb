```python
import numpy as np
import matplotlib.pyplot as plt

import dmsh
import meshio
```

### Cerchio

```python
#
geo = dmsh.Circle([0.0, 0.0], 1.0)
X, cells = dmsh.generate(geo, 0.2)

#
plt.figure ( 0 , figsize = ( 5 , 5 ) )
plt.scatter(X[:,0], X[:,1])
plt.grid(True)
plt.show()

# visualize the mesh
dmsh.helpers.show(X, cells, geo)
```

### Ciambella

```python
#
geo = dmsh.Difference(dmsh.Circle([0, 0.0], 1.0), dmsh.Circle([0, 0.0], 0.5))
X, cells = dmsh.generate(geo, 0.1)

#
plt.figure ( 0 , figsize = ( 5 , 5 ) )
plt.scatter(X[:,0], X[:,1])
plt.grid(True)
plt.show()

# visualize the mesh
dmsh.helpers.show(X, cells, geo)
```

```python
#
r = dmsh.Rectangle(-1.0, +1.0, -1.0, +1.0)
c = dmsh.Circle([0.0, 0.0], 0.3)
geo = dmsh.Difference(r, c)

X, cells = dmsh.generate(geo, lambda pts: np.abs(c.dist(pts)) / 5 + 0.05, tol=1.0e-10)

#
plt.figure ( 0 , figsize = ( 5 , 5 ) )
plt.scatter(X[:,0], X[:,1])
plt.grid(True)
plt.show()

# visualize the mesh
dmsh.helpers.show(X, cells, geo)
```
